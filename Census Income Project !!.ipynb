<Strong><B> DataSet : <Strong><B>

The data we have consists of 32,560 entries and includes 14 separate features that help describe each entry. Our goal is to forecast whether a person earns more than $50k per year or not.

Because the predictions fall into two categories (>50K or <=50K), this problem is a classification task. We will use classification models to train and make predictions about whether a person's income exceeds $50,000 annually.

Here are the details of the features we have, and we will input them into our classification model for training.

1. Age — The age of an individual, this ranges from 17 to 90.

2. Workclass — The class of work to which an individual belongs.

3. Fnlwgt — The weight assigned to the combination of features (an estimate of how many people belong to this set of combination)

4. Education — Highest level of education

5. Education_num — Number of years for which education was taken

6. Marital_Status — Represents the category assigned on the basis of marriage status of a person

7. Occupation — Profession of a person

8. Relationship — Relation of the person in his family

9. Race — Origin background of a person

10. Sex — Gender of a person

11. Capital_gain — Capital gained by a person

12. Capital_loss — Loss of capital for a person

13. Hours_per_week — Number of hours for which an individual works per week

14. Native_Country — Country to which a person belongs

Output:

1. Annual_income — The target variable, which predicts if the income is higher or lower than 50K$.

### Outlier Detection:
#### *Inference: Nearly all the columns are needed to be treated, since each one of them are having outliers.*

```python
df.columns
```

```python
for col in df.columns:
    if df[col].dtype != "object" :
        plt.boxplot(df[col])
        plt.xlabel(col)
        plt.show()
```

```python
out_col = ['hours-per-week', 'age', 'fnlwgt', 'education-num']
# treatment
for col in out_col:
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  # Define my IQR
  IQR = Q3 - Q1
  # Defining my limits / whiskers
  LW = Q1 - 1.5*IQR  # lower limit / lower whisker
  UW = Q3 + 1.5*IQR
  # formulation
  df = df[(df[col]>=LW) & (df[col]<=UW)]
```

```python
out_cols = ['hours-per-week', 'age', 'fnlwgt', 'education-num']

for col in out_cols:
    if df[col].dtype != "object":
        plt.boxplot(df[col])
        plt.xlabel(col)
        plt.title(f'Boxplot for {col}')
        plt.show()
```

### Scaling the data:

The next step is to bring the data to a common scale, since there are certain columns with very small values and some columns with high values. This process is important as values on a similar scale allow the model to learn better.

We use standard scaler for this process –

‘StandardScaler follows Standard Normal Distribution (SND). Therefore, it makes mean = 0 and scales the data to unit variance’

```python
from sklearn.preprocessing import StandardScaler

# where x is the feature matrix and y is the target variable
x = df.iloc[:, :-1]
y = df['annual_income']
# Initialize the StandardScaler
sc = StandardScaler()
# Fit and transform the features
x_scaled = sc.fit_transform(x)
# Now x_scaled contains the scaled features
x_scaled
```

## ***Fitting the data into classification models:***

***We now proceed to the main step of our machine learning, fitting the model and predicting the outputs. We fit the data into multiple classification models to compare the performance of all models and select the best model***

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
#splitting the data 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=0)
```

```python
x_train.shape, y_test.shape
```

### *LOGISTIC REGRESSION:*

```python
from sklearn.linear_model import *
clf = LogisticRegression()
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
print(f"The overall accuracy of the data on performing LR: {accuracy_score(y_pred, y_test)*100}")
```

## *Decision Tree:*

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
# train
clf.fit(x_train, y_train)
# test
pred = clf.predict(x_test)
print(f"The overall accuracy of the data on performing Decision Tree: {accuracy_score(pred, y_test)*100}")
```

```python
print(confusion_matrix(pred, y_test))
```

## *Random Forest Classifier:*

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
# train
clf.fit(x_train, y_train)
# test
pred = clf.predict(x_test)
print(f"The overall accuracy of the data on performing Random Forest Classifier: {accuracy_score(pred, y_test)}")
```

```python
print(confusion_matrix(pred, y_test))
```

```python
print(classification_report(pred, y_test))
```

### *Summary of the classification Report:*

##### *Precision: Precision is a measure of the accuracy of the positive predictions.

* For class 0 (<=50K), the precision is 92%, indicating that when the model predicts an individual's income to be <=50K, it is correct 92% of the time.

* For class 1 (>50K), the precision is 63%, meaning that when the model predicts an individual's income to be >50K, it is correct 63% of the time.

##### *Recall: Recall, also known as sensitivity or true positive rate, measures the ability of the model to capture all relevant instances.

* For class 0, the recall is 87%, indicating that the model correctly identifies 87% of all individuals with actual income <=50K.
* For class 1, the recall is 72%, meaning that the model correctly identifies 72% of all individuals with actual income >50K.
* F1-score: The F1-score is the harmonic mean of precision and recall, providing a balance between the two metrics.

* For class 0, the F1-score is 90%.
* For class 1, the F1-score is 67%.

#### *Accuracy: Overall accuracy of the model is 84%, meaning that the model correctly predicts the income class for 84% of the individuals.

#### *Macro Avg: The macro average of precision, recall, and F1-score is around 77-80%.

#### *Weighted Avg: The weighted average of precision, recall, and F1-score is around 84%, considering the class imbalance.
